```python
# -*- coding: utf-8 -*-
"""

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1ad6xp0KYJeE_x2pF1Ze8w9iBQuCKcFko


#**Import packages**
"""

import math
from operator import itemgetter

#%%
# Pearson Correlation between two vectors
# Note: the formula has been modified to return the pc on a scale of 0 to 1 (instead of -1 to 1)

# Pearson Correlation between two vectors
def pearsonD(userratings1, userratings2, debug = "N"):

    sumpq = 0
    sump = 0
    sumq = 0
    sump2 = 0
    sumq2 = 0

    # calcualte n
    n = len (userratings1.keys() & userratings2.keys())

    # error check for n==0 condition
    if n == 0:
        print (">>> Error! No common keys (n=0). Returning -2!") if debug == "Y" else 0
        return -2

    # calculate pearson correlation using the computationally efficient form
    for item in (userratings1.keys() & userratings2.keys()):
        # consider item rating only if both users have rated item
        p = userratings1[item]
        q = userratings2[item]
        sumpq += p * q
        sump += p
        sumq += q
        sump2 += pow(p, 2)
        sumq2 += pow(q, 2)

    # calcualte nr and dr for pearson correlation
    nr = (sumpq - (sump * sumq) / n)

    dr = (math.sqrt(sump2 - pow(sump, 2) / n) *
                        math.sqrt(sumq2 - pow(sumq, 2) / n))

    # error check for dr==0 condition
    if dr == 0:
        print (">>> Error! Denominator is 0. Returning -2!") if debug == "Y" else 0
        return -2

    # calculate pearson correlation coefficient
    r = nr / dr

    # it is customary to convert pearson correlation to a scale of 0 to 1 (instead of -1 to 1) using the below formula
    r = (r + 1) / 2

    # return value of pearson correlation coefficient
    return r

#%%

#%%
# user ratings
SongData = {"Angelica": {"Blues Traveler": 3.5, "Broken Bells": 2.0, "Norah Jones": 4.5, "Phoenix": 5.0, "Slightly Stoopid": 1.5, "The Strokes": 2.5, "Vampire Weekend": 2.0},
         "Bill":{"Blues Traveler": 2.0, "Broken Bells": 3.5, "Deadmau5": 4.0, "Phoenix": 2.0, "Slightly Stoopid": 3.5, "Vampire Weekend": 3.0},
         "Chan": {"Blues Traveler": 5.0, "Broken Bells": 1.0, "Deadmau5": 1.0, "Norah Jones": 3.0, "Phoenix": 5, "Slightly Stoopid": 1.0},
         "Dan": {"Blues Traveler": 3.0, "Broken Bells": 4.0, "Deadmau5": 4.5, "Phoenix": 3.0, "Slightly Stoopid": 4.5, "The Strokes": 4.0, "Vampire Weekend": 2.0},
         "Hailey": {"Broken Bells": 4.0, "Deadmau5": 1.0, "Norah Jones": 4.0, "The Strokes": 4.0, "Vampire Weekend": 1.0},
         "Jordyn":  {"Broken Bells": 4.5, "Deadmau5": 4.0, "Norah Jones": 5.0, "Phoenix": 5.0, "Slightly Stoopid": 4.5, "The Strokes": 4.0, "Vampire Weekend": 4.0},
         "Sam": {"Blues Traveler": 5.0, "Broken Bells": 2.0, "Norah Jones": 3.0, "Phoenix": 5.0, "Slightly Stoopid": 4.0, "The Strokes": 5.0},
         "Veronica": {"Blues Traveler": 3.0, "Norah Jones": 5.0, "Phoenix": 4.0, "Slightly Stoopid": 2.5, "The Strokes": 3.0}
        }

# define "k" for KNN -- number of nearest neighbors to consider for recommendation
k = 3

# for whom are we making recommendations?
# Prompt to enter the username for whom we need to provide recommendations
# Prompt to re-enter if the username is not in SongData
while True:
  print(f"Enter the user who needs recommedations:")
  print(f"(Eg: {', '.join(list(SongData.keys()))})")
  userX = input("    ")
  if userX not in SongData.keys():
    print(f"Sorry: {userX} is not in our database.  Please try again.")
  else:
    break

# Check if the code should run in DEBUG mode
# Prompt to re-enter if the input is not Y or y or N or n
while True:
  print("Enter Y if you want to run in DEBUG mode, else N:")
  debug = input("    ")
  debug = debug.upper() # convert to UPPER case
  if debug not in ('Y', 'N'):
    print("Sorry: You must enter either Y or N.  Please try again.")
  else:
    break

# Extract ratings given for different albums by userX
userXRatings = SongData[userX]

print("----------------------------------------------")
print(f"Working on recommendations for userX = {userX}")
print(f"DEBUG mode: {debug}")
print()
if debug == 'Y':
  print("userXRatings = ")
  for key in userXRatings.keys():
    print(f"{key}:{userXRatings[key]}")
print()

#%%

"""#**Step 1**
*Find the pearson correlation distances between userX's ratings, and each of the other user's ratings.*
"""

#%%
# (1) find the pearson correlation distances between userX's ratings, and each of the other user's ratings.
# use a for loop to get at the other users and their ratings - DO NOT hard code.
# use the given function to calculate the pearson correlation
# assign list of (user, distance) tuples to variable userDistances.
#       Note: Do not include users whose pearson correlation is UNDEFINED (i.e. if pc = -2)
# Example: If userX is "Veronica", userDistances should be: [('Angelica', 0.91), ('Bill', 0.12), ('Chan', 0.64), ('Dan', 0.12), ('Jordyn', 0.87), ('Sam', 0.22)]

userDistances = []
# <<<<< (1) YOUR CODE STARTS HERE >>>>>
for otherUser in SongData:
    if otherUser != userX:  # skip the target user
        pc = pearsonD(userXRatings, SongData[otherUser])
        if pc != -2:  # ignore undefined correlations
            userDistances.append((otherUser, round(pc, 2)))


# <<<<< (1) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print()
  print(f"userDistances (before sorting):")
  for userDistance in userDistances:
    print(userDistance)


#%%

"""#**Step 2**
*Sort list of tuples by highest distance to lowest distance (since higher the correlation greater the similarity)*
"""

#%%
#(2) Sort list of tuples by highest distance to lowest distance (since higher the correlation greater the similarity).
# assign sorted list to variable userSortedDistances.
# Example: If userX is "Veronica", userSortedDistances should be: [('Angelica', 0.91), ('Jordyn', 0.87), ('Chan', 0.64), ('Sam', 0.22), ('Bill', 0.12), ('Dan', 0.12)]

userSortedDistances = []

# <<<<< (2) YOUR CODE STARTS HERE >>>>>

userSortedDistances = sorted(userDistances, key=lambda x: x[1], reverse=True)


# <<<<< (2) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print(f"{userSortedDistances} (after sorting):")
  for userDistance in userSortedDistances:
    print(userDistance)

#%%

"""#**Step 3**

*Determine userX's k Nearest Neighbors.  They are the k users from the 0th position of the sorted list.*

**Step 3.1**
"""

#%%
# (3) userX's k Nearest Neighbors are the k users from the 0th position of the sorted list.
# (3.1)
# Extract first k elements from userSortedDistances and assign to a variable userKNNDistances
# Example: If userX is "Veronica", userKNNDistances should be: [('Angelica', 0.91), ('Jordyn', 0.87), ('Chan', 0.64)]

userKNNDistances = []
# <<<<< (3.1) YOUR CODE STARTS HERE >>>>>

userKNNDistances = userSortedDistances[:k]


# <<<<< (3.1) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print(f"KNN for {userX} and their distances are:")
  for userDistance in userKNNDistances:
    print(userDistance)

#%%

"""**Step 3.2**"""

#%%
# (3.2)
# Extract KNN user (names) from userKNNDistances
# Example: If userX is "Veronica", userKNN should be: ['Angelica', 'Jordyn', 'Chan']
# <<<<< (3.2) CODE ALREADY GIVEN HERE.  You do not need to code anything for 3.2 >>>>>

if debug == 'Y':
  userKNN = [user for (user, distance) in userKNNDistances]
  print(f"KNN for {userX} are:")
  for user in userKNN:
    print(user)

#%%

"""#**Step 4**

*Compute the weights for the nearest neighbors based on their distances (pearson correlation) with userX*

**Step 4.1**
"""

#%%
# (4) compute the weights for the nearest neighbors based on their distances (pearson correlation) with userX
# weight of one Nearest Neighbor (NN) = (distance of that NN with userX) / (sum of distances between each NN with userX)

# (4.1) first compute the DENOMINATOR for computing weights - which is sum of the distances of K NNs with userX
# initialize a variable sumKNNDistances to 0
# use a for loop to iterate over userKNNDistances
#     and compute the cumulative sum of the distances (of each NN with userX) in the variable sumKNNDistances
# Note: userKNNDistances contains tuples.  Hence, use the following format for the for loop:
#     for (user, distance) in userKNNDistances:
# Example: If userX is "Veronica", sumKNNDistances should be: 2.42

sumKNNDistances = 0

# <<<<< (4.1) YOUR CODE STARTS HERE >>>>>
for (user, distance) in userKNNDistances:

    sumKNNDistances += distance


# <<<<< (4.1) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print(f"sumKNNDistances = {sumKNNDistances}")


#%%

"""**Step 4.2**"""

#%%
# (4.2) Next, determine the weight for each NN and store them in a dictionary variable userKNNWeights.
# use a for loop to iterate over userKNNDistances to compute the weight of each nearest user
# KEYS in the dictionary userKNNWeights will be the users and the VALUES will be the weights
# Example: If userX is "Veronica", userKNNWeights should be: {'Angelica': 0.38, 'Jordyn': 0.36, 'Chan': 0.26}

userKNNWeights = {}

# <<<<< (4.2) YOUR CODE STARTS HERE >>>>>

for (user, distance) in userKNNDistances:
    weight = distance / sumKNNDistances
    userKNNWeights[user] = round(weight, 2)


# <<<<< (4.2) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print()
  print(f"userKNNWeights are {userKNNWeights}")

#%%

"""#**Step 5**

*Prepare recommendations for userX*

**Step 5.1**
"""

#%%
# (5) Recos for userX will include albums rated by ALL userKNNs, BUT not rated by userX.
# (5.1) Determine the potential albums for recommendations
# use a for loop to iterate over userKNNDistances (or userKNNWeights.  Either one works)...
#     determine the albums rated by each nearest user but not rated by userX
#     append these albums to a list itemsToRecommend
#     Note: you may need to use itemsToRecommend.extend(...) instead of itemsToRecommend.append(...) to append a "list" to a list.
# Ensure that you consider an album only once even if it was rated by multple users
# Example: If userX is "Veronica", itemsToRecommend (after removing duplicates) should be: {'Broken Bells', 'Vampire Weekend', 'Deadmau5'}

itemsToRecommend = []
# <<<<< (5.1) YOUR CODE STARTS HERE >>>>>

for neighbor, distance in userKNNDistances:
    neighborAlbums = SongData[neighbor]

    for album in neighborAlbums:
        if album not in userXRatings:
            itemsToRecommend.append(album)

itemsToRecommend = list(set(itemsToRecommend))



# <<<<< (5.1) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print(f"Potential items (after removing duplicates) for recommendation are:")
  for item in itemsToRecommend:
    print(item)

#%%

"""**Step 5.2**"""

#%%
# (5.2)
# Compute the Predicted Rating of userX for each album in the potential recommendation list
#           ...using the weights and ratings of each NN
# Assign list of (album, predictedRating) tuples to variable userXRecos.
# Example: If userX is "Veronica", userXRecos should be: [('Broken Bells', 2.64), ('Vampire Weekend', 2.2), ('Deadmau5', 1.7)]

userXRecos = []

for album in itemsToRecommend:
    weighted_sum = 0

    for (neighbor, distance) in userKNNDistances:
        neighborRatings = SongData[neighbor]

        if album in neighborRatings:
            weighted_sum += userKNNWeights[neighbor] * neighborRatings[album]

    userXRecos.append((album, round(weighted_sum, 2)))


# <<<<< (5.2) YOUR CODE ENDS HERE >>>>>

if debug == 'Y':
  print()
  print(f"Recommendations (before sorting):")
  for reco in userXRecos:
    print(reco)

#%%

"""#**Step 6**

*Provide the recommendations for userX sorted by predicted rating*
"""

#%%
# (6) Determine the recommendations for userX
# Sort list of tuples by highest predicted rating to lowest predicted rating.
# assign sorted list to varaible userXSortedRecos.
# Example: If userX is "Veronica", userXSortedRecos should be: [('Broken Bells', 2.64), ('Vampire Weekend', 2.2), ('Deadmau5', 1.7)]


# <<<<< (6) YOUR CODE STARTS HERE >>>>>

userXSortedRecos = sorted(userXRecos, key=lambda x: x[1], reverse=True)

# <<<<< (6) YOUR CODE ENDS HERE >>>>>

print (f"Here are the recommendations for {userX}:")
print ("------------------------------------------")
for reco in userXSortedRecos:
  print(reco)

#%%
```
